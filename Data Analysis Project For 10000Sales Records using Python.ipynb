# 1.Business_Understanding
## 1. Business Problem
The dataset contains sales transaction records for a multinational retail business that operates across various regions and product categories. The business faces challenges in:

Identifying trends in sales performance across different regions, product types, and order channels.
Understanding profitability variations based on unit price, unit cost, and total revenue.
Optimizing supply chain efficiency by analyzing shipping dates and order priorities.

## 2. Business Goal
The goal of this analysis is to derive actionable insights to improve strategic decision-making in the following areas:

Sales Performance Analysis: Identify top-performing regions, countries, and product categories.
Profitability Insights: Understand how different products contribute to overall profits and where cost optimizations can be made.
Operational Efficiency: Evaluate shipping timelines, order priorities, and their impact on sales performance.
Sales Channel Effectiveness: Compare online vs. offline sales to determine the best-performing channel.

## 3. Data Overview
The dataset consists of 10,000 sales records with the following key attributes:

Transaction Details: Order ID, Order Date, Ship Date, Sales Channel, and Order Priority.
Geographical Information: Region and Country.
Product Details: Item Type (e.g., Office Supplies, Beverages, Vegetables).
Financial Metrics: Units Sold, Unit Price, Unit Cost, Total Revenue, Total Cost, and Total Profit.

# 4.Modeling

```python
x=df.drop('Total Revenue',axis=1)
y=df['Total Revenue']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
y_predict_rf=rf.predict(x_test)
y_predict_dt=dt.predict(x_test)
```

# 5.Evaluation

```python
print('MAE',mean_absolute_error(y_test,y_predict_rf))
print('MAE',mean_absolute_error(y_test,y_predict_dt))
```

```python
print('MSE',mean_squared_error(y_test,y_predict_rf))
print('MSE',mean_squared_error(y_test,y_predict_dt))
```

```python
print('r2',r2_score(y_test,y_predict_rf))
print('r2',r2_score(y_test,y_predict_dt))
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
grid_pram={'max_depth':[10,20,7],'max_leaf_nodes':[20,40,30]}
grid_model=GridSearchCV(DecisionTreeRegressor(),grid_pram,cv=5)
grid_model.fit(x_train,y_train)
```

```python
print(grid_model.best_params_)
```

```python
y_predict_grid=grid_model.best_estimator_.predict(x_test)
```

```python
print('MAE',mean_absolute_error(y_test,y_predict_grid))
print('MSE',mean_squared_error(y_test,y_predict_grid))
print('r2',r2_score(y_test,y_predict_grid))
```

```python
ron_pram={'max_depth':[10,20,25],'n_estimators':[10,20,30]}
ron_model=RandomizedSearchCV(estimator=RandomForestRegressor(),param_distributions=ron_pram,cv=5)
ron_model.fit(x_train,y_train)
```

```python
print(ron_model.best_params_)
```

```python
y_predict_ron=ron_model.best_estimator_.predict(x_test)
```

```python
print('MAE',mean_absolute_error(y_test,y_predict_ron))
print('MSE',mean_squared_error(y_test,y_predict_ron))
print('r2',r2_score(y_test,y_predict_ron))
```

```python
import gradio as gr
```

```python
def prediction_sales_Revenue(RE,IT,SC,OP,US,UP,UC,TC,TP):
    try:
        input_data=pd.DataFrame({
            'Region':[RE],
            'Item Type':[IT],
            'Sales Channel':[SC],
            'Order Priority':[OP],
            'Units Sold':[US],
            'Unit Price':[UP],
            'Unit Cost':[UC],
            'Total Cost':[TC],
            'Total Profit':[TP]
        })
        for col in le_col:
            input_data[col]=le[col].transform(input_data[col])
        input_data[scaler_col]=scaler.transform(input_data[scaler_col])
        prediction=rf.predict(input_data)
        return prediction[0]
    except Exception as e:
        return str(e)
gr.Interface(
    inputs=[
        gr.Dropdown(['Europe','Sub-Saharan Africa','Asia','Middle East and North Africa',
                     'Central America and the Caribbean','Australia and Oceania','North America'],label='Region'),
        gr.Dropdown(['Personal Care','Household','Clothes','Baby Food','Office Supplies','Vegetables',
                     'Cosmetics','Cereal','Snacks','Meat','Fruits','Beverages'],label='Item Type'),
        gr.Dropdown(['Online','Offline'],label='Sales Channel'),
        gr.Dropdown(['C','H','L','M'],label='Order Priority'),
        gr.Number(label='Units Sold'),
        gr.Number(label='Unit Price'),
        gr.Number(label='Unit Cost'),
        gr.Number(label='Total Cost'),
        gr.Number(label='Total Profit')
    ],
    fn=prediction_sales_Revenue,
    outputs=gr.Textbox(label='Prediction')
).launch()
```
